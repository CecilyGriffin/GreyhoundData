## Racing Data for PJ Cheap Thrills

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
datapath = "Resources/PJRaceHistory.csv"

# Read the mouse data and the study results
racedata = pd.read_csv(datapath)
```

```python
# convert to a dataframe
racedatadf = pd.DataFrame(racedata)
racedatadf
```

```python
dateplace = racedata[['Date','Fin']]
#dateplace.info()
dateplace['Date']= pd.to_datetime(dateplace['Date'])
sortdp = dateplace.sort_values(by='Date',ascending=True)
sortdp.head(10)
```

```python
sortdp['Date'] = sortdp.Date.astype(str)
sortdp = sortdp.replace('1st', 1).replace('2nd', 2).replace('3rd', 3).replace('4th', 4).replace('5th', 5).replace('6th', 6).replace('7th', 7).replace('8th', 8)
```

```python
#plot a linegraph
plt.plot(sortdp['Date'], sortdp['Fin']) 
plt.xlabel('Date')
plt.ylabel('Standing')
plt.title('Career Standings for PJ Cheap Thrills')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```
